### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Mostrar estructura de la primera tienda
print("Vista previa de tienda 1:")
display(tienda.head())
```

```python
# Calcular ingresos totales (sumar columna 'Precio')
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()
```

```python
ingreso_t1
```

```python
ingreso_t2
```

```python
ingreso_t3
```

```python
ingreso_t4
```

```python
# Crear resumen general
ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingreso Total": [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]
})
ingresos
```

```python
# Formatear a 2 decimales
ingresos["Ingreso Total"] = ingresos["Ingreso Total"].round(2)
ingresos
```

```python
# Mostrar resultado ordenado
display(ingresos.sort_values("Ingreso Total", ascending=False))
```

```python
#Gráfico de comparación entre tiendas
import matplotlib.cm as cm
import numpy as np

colors = cm.Greens(np.linspace(0.4, 0.9, len(ingresos)))

plt.figure(figsize=(6,4))
plt.bar(ingresos["Tienda"], ingresos["Ingreso Total"], color=colors, edgecolor="black")
plt.title("Ingresos totales por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso total ($)")

# Etiquetas de valor encima de cada barra
for i, v in enumerate(ingresos["Ingreso Total"]):
    plt.text(i, v + (v * 0.01), f"${v:,.0f}", ha="center", fontsize=9)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt
```

```python
# Agregar columna 'Tienda' a cada DataFrame
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"
```

```python
# Combinar todas las tiendas en un solo DataFrame
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])
todas_tiendas
```

```python
# Agrupar por Tienda y Categoría, contar número de ventas
ventas_categoria = (
    todas_tiendas.groupby(["Tienda", "Categoría del Producto"])
    .size()
    .reset_index(name="Cantidad de Ventas")
)
ventas_categoria
```

```python
# Mostrar la tabla completa
display(ventas_categoria.head(10))
```

```python
top_categorias = (
    ventas_categoria.sort_values(["Tienda", "Cantidad de Ventas"], ascending=[True, False])
    .groupby("Tienda")
    .head(5)
)

display(top_categorias)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(
    data=ventas_categoria,
    x="Categoría del Producto",
    y="Cantidad de Ventas",
    hue="Tienda",
    palette="Set2"
)
plt.title("Comparación de ventas por categoría y tienda")
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)
plt.legend(title="Tienda", bbox_to_anchor=(1.02,1), loc="upper left")
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular promedio de calificación por tienda
valoracion_media = (
    todas_tiendas.groupby("Tienda")["Calificación"]
    .mean()
    .reset_index()
    .rename(columns={"Calificación": "Valoración Promedio"})
)
valoracion_media
```

```python
# Redondear a dos decimales
valoracion_media["Valoración Promedio"] = valoracion_media["Valoración Promedio"].round(2)

# Mostrar tabla
display(valoracion_media)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(valoracion_media["Valoración Promedio"], valoracion_media["Tienda"],
            color="teal", s=150, alpha=0.7, edgecolors="black")

plt.title("Valoración promedio por tienda")
plt.xlabel("Calificación promedio (1–5)")
plt.xlim(0, 5)

# Etiquetas de valor al lado del punto
for i, row in valoracion_media.iterrows():
    plt.text(row["Valoración Promedio"] + 0.05, row["Tienda"], f"{row['Valoración Promedio']:.2f}",
             va="center", fontsize=9)

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Agrupar por tienda y producto, contar cantidad de ventas
ventas_productos = (
    todas_tiendas.groupby(["Tienda", "Producto"])
    .size()
    .reset_index(name="Cantidad de Ventas")
)
ventas_productos
```

```python
# Top 5 productos más vendidos por tienda
top5_productos = (
    ventas_productos
    .sort_values(["Tienda", "Cantidad de Ventas"], ascending=[True, False])
    .groupby("Tienda")
    .head(5)
)

display(top5_productos)
```

```python
# 5 productos menos vendidos
bottom5_productos = (
    ventas_productos
    .sort_values(["Tienda", "Cantidad de Ventas"], ascending=[True, True])
    .groupby("Tienda")
    .head(5)
)

display(bottom5_productos)
```

```python
# Gráfico Top 5 más vendidos por tienda
for tienda in top5_productos["Tienda"].unique():
    datos_top = top5_productos[top5_productos["Tienda"] == tienda]
    plt.figure()
    plt.barh(datos_top["Producto"], datos_top["Cantidad de Ventas"], color="mediumseagreen")
    plt.title(f"Top 5 productos más vendidos - {tienda}")
    plt.xlabel("Cantidad de Ventas")
    plt.ylabel("Producto")
    plt.gca().invert_yaxis()  # para que el más vendido quede arriba
    plt.tight_layout()
    plt.show()
```

```python
# Bottom 5 productos menos vendidos

for tienda in bottom5_productos["Tienda"].unique():
    datos_bottom = bottom5_productos[bottom5_productos["Tienda"] == tienda]
    plt.figure()
    plt.barh(datos_bottom["Producto"], datos_bottom["Cantidad de Ventas"], color="salmon")
    plt.title(f"Bottom 5 productos menos vendidos - {tienda}")
    plt.xlabel("Cantidad de Ventas")
    plt.ylabel("Producto")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio_promedio = (
    todas_tiendas.groupby("Tienda")["Costo de envío"]
    .mean()
    .reset_index()
    .rename(columns={"Costo de envío": "Costo de Envío Promedio"})
)
envio_promedio
```

```python
# Redondear a dos decimales
envio_promedio["Costo de Envío Promedio"] = envio_promedio["Costo de Envío Promedio"].round(2)


display(envio_promedio)
```

```python
# Gráfico del valor de envío promedio por tienda

plt.figure(figsize=(6,4))
plt.fill_between(envio_promedio["Tienda"], envio_promedio["Costo de Envío Promedio"],
                 color="navajowhite", alpha=0.6)
plt.plot(envio_promedio["Tienda"], envio_promedio["Costo de Envío Promedio"],
         color="darkorange", linewidth=2, marker="o", markersize=8, markeredgecolor="black")

plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de envío promedio ($)")
plt.grid(True, linestyle="--", alpha=0.4)

for i, v in enumerate(envio_promedio["Costo de Envío Promedio"]):
    plt.text(i, v + 0.05, f"${v:.2f}", ha="center", fontsize=9, color="black")

plt.tight_layout()
plt.show()
```
